```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# 1. Definir Transformaciones (Preprocessing)
# Normalizamos a un rango de [-1, 1] para mejorar la convergencia
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# 2. Descargar y Cargar Dataset
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                         shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 
           'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
# Función para mostrar imágenes
def imshow(img):
    img = img / 2 + 0.5     # Des-normalizar para visualizar correctamente
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# Obtener algunas imágenes aleatorias
dataiter = iter(trainloader)
images, labels = next(dataiter)

# Mostrar imágenes y dimensiones
print(f"Image Batch Shape: {images.shape}") # Debería ser [4, 3, 32, 32]
print(f"Labels: {[classes[labels[j]] for j in range(4)]}")
imshow(torchvision.utils.make_grid(images))

# Análisis de Distribución (Brief analysis)
total_samples = len(trainset)
print(f"Total training samples: {total_samples}")
print(f"Image dimensions: {images[0].shape[1]}x{images[0].shape[2]} pixels")
print(f"Channels: {images[0].shape[0]} (RGB)")
```
